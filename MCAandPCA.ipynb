# Multiple Correspondence Analysis and Principal Component Analysis

This Notebook describes the process of obtaining, through Multiple Correspondence Analysis and Principal Component Analysis, a ranking (which will be compared to the price variable) of a dataset of houses (which contains numeric and categorical variables). Using the MCA method, the coordinates concerning the categorical variables will be found, and then, together with the numerical ones will be the input for the PCA, which will decrease its number of parameters and generate the ranking.

### Methods Description

- **Multiple Correspondence Analysis (MCA)** is a data analysis technique for nominal categorical data that allows finding the association between two or more qualitative variables. Steps:

     - Check the correspondence between variables (variables with no significant correspondence wouldn't be used);
     
     - Calculate the number of dimentions;
     
     - Get the coordinates for using it as input.


- **Principal Component Analysis (PCA)** is a dimensionality-reduction method that is often used to reduce the dimensionality of large data sets, by transforming a large set of variables into a smaller one that still contains most of the information in the large set. In this case, it will also provide a ranking, based on the factors and their variance. Steps:

    - Normalize the numeric variables;
    
    - Check, using the Bartlett's Test of Sphericity if the PCA can be used.
    
    - Find the eigenvalues;
    
    - Check, using the Kaiser's Test the ideal number of components;
    
    - Get the ranking, by component x it's explained variance.

## Multiple Correspondence Analysis

Multiple Match Analysis is based on the correspondence between columns. If there is a column with no significant correspondence to any other column it can be removed, the test uses the chi-square contingency value.

If the minimum match of a column to another column is less than 0.05, the column is retained, otherwise it will be dropped.

```python
# Testing the minimum correspondence between columns
assosc_table = pd.DataFrame(index=df_houses_cat.columns, columns=df_houses_cat.columns)
for idx in assosc_table.index:
    for col in assosc_table.columns:
        if col != idx:
            contingency = df_houses_cat[[col, idx]].value_counts().reset_index().pivot(index=[col], columns=[idx])[0].fillna(0)
            assosc_table[col].loc[idx] = scipy.stats.chi2_contingency(contingency)[1]
assosc_table.min()
```

```python
# Checking association table
assosc_table
```

All variables are statistically associated with at least one other variable. It is observed that 'water_front' has a weak correspondence with all variables except 'large_living_room'.

## Principal Component Analysis

```python
# Checking numeric variables
df_houses_num.head(5)
```

Once the ranking will target the 'property_value', the price variable will be discarded.

```python
# Dropping 'property_value'
df_houses_num = df_houses_num.drop('property_value', 1)
```

```python
# Merging MCA coordinates and numeric variables
df_houses_total = pd.merge(df_houses_num, mca, how='inner', left_index=True, right_index=True)
```

```python
# Normalizing parameters
scaler = StandardScaler()
df_houses_normalized = pd.DataFrame(data=scaler.fit_transform(df_houses_total))
```

The PCA methods is based on the correlation between the parameters.

The Bartlett's Test of Sphericity compares an observed correlation matrix to the identity matrix, if the p-value < 0.05, the parameters are not correlated, so there's no point on applying the PCA.

```python
# Bartlett's Test of Sphericity
chi_square_value, p_value = calculate_bartlett_sphericity(df_houses_normalized)
if p_value < 0.05:
    print('PCA can be applied.')
else:
    print('PCA cannot be applied.')
```

```python
# Correlation matrix
df_corr = df_houses_normalized.corr()
df_corr
```

The Kaiser Rule defines that only eigenvalues greater than 1 will generate factors.

```python
# Defining number of values
v, w = np.linalg.eig(df_corr.values.astype(float))
n_comp = (v>1).sum()
print(f'Applying the Kaiser Rule, {n_comp} factors will be used.')
```

```python
# Applying PCA
pca = PCA(n_components = 4, whiten=True)
df_houses_factors = pd.DataFrame(pca.fit_transform(df_houses_normalized))
df_houses_factors.head(5)
```

Once the factors were obtained - for the 4953 houses - the ranking will be gotten by this factors and the explained variance correspondent to them.

```python
# Explained variances
pca.explained_variance_
```

```python
# Checking the weights of the variables on the components 
pd.DataFrame(pca.components_, columns = df_houses_total.columns)
```

The first is the most relevant component, by far. And the variables mor relevant for this factor is the 'land_size_sqm',	'house_size_sqm', 'no_of_rooms', 'no_of_bathrooms' and the mca first component [0] (in a opposite way, once the weight is negative). 

```python
# Adding the ranking to the original dataset
df_houses['ranking'] = (pca.explained_variance_[0] * df_houses_factors[0]) + (pca.explained_variance_[1] * df_houses_factors[1]) +  (pca.explained_variance_[2] * df_houses_factors[2]) + (pca.explained_variance_[3] * df_houses_factors[3])
df_houses.head(5)
```

```python
# Plotting registers property value and ranking
plt.figure(figsize=(8,6))
plt.scatter(df_houses['property_value'], df_houses['ranking'], alpha=.2, s=20)
plt.xlabel('Property value')
plt.ylabel('Ranking')
plt.title('Property value vs PCA Ranking')
plt.show()
```

It's clearly noticed by the Plot that there's a strong correlation between the Ranking gotten (and, for get it, the property value wasn't used at all) and the property value.
